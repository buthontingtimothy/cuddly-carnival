```python
aapl = pd.read_csv("aapl.csv")
```

```python
aapl
```

```python
aapl['Date'] = pd.to_datetime(aapl['Date'], utc=True)
```

```python
aapl['CumMax'] = aapl['Close'].cummax()
```

```python
aapl['DrawDown'] = 1 - aapl['CumMax']/ aapl['Close']
```

```python
aapl[aapl.DrawDown == aapl.DrawDown.min()]['CumMax']
```

```python
aapl[aapl.DrawDown == aapl.DrawDown.min()]['CumMax'].values[0]
```

```python
aapl[aapl.CumMax == aapl[aapl.DrawDown == aapl.DrawDown.min()]['CumMax'].values[0]]
```

```python
aapl[aapl['CumMax'] >= 100].sort_values('DrawDown')
```

```python
aapl[aapl['CumMax'] >= 150].sort_values('DrawDown')
```

```python
aapl[aapl['CumMax'] >= 200].sort_values('DrawDown')
```

```python
gp_cummax = aapl[['Date','CumMax']].groupby('CumMax')
```

```python
day_recover = gp_cummax.min().join(gp_cummax.max(), lsuffix="_min", rsuffix="_max")
```

```python
day_recover['recover'] = day_recover['Date_max'] - day_recover['Date_min']
```

```python
day_recover[day_recover['recover'] == day_recover['recover'].max()]
```

```python
day_recover[day_recover.index >= 100].sort_values('recover')
```

```python
day_recover[day_recover.index >= 150].sort_values('recover')
```

```python
day_recover[day_recover.index >= 200].sort_values('recover')
```
